#Diabetic Prediction
**The aim of this project is to determine whether a person has diabetics or not.**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/siddhardhan23/multiple-disease-prediction-streamlit-app/main/dataset/diabetes.csv")
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['Outcome'].value_counts()
```

```python
x = df.drop(['Outcome'], axis='columns')
y = df['Outcome']
```

```python
x.head()
```

```python
y.head(3)
```

**Splitting the data into Train and Test**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
len(x_train)
```

```python
len(x_test)
```

**Model Building**

```python
from sklearn import svm
model = svm.SVC(kernel ='linear' , C=1 , gamma='auto')
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

**Predict Diabetics for input data**

```python
input_data = (5,166,72,19,175,25.8,0.587,51)
input_data_as_np = np.array(input_data)
input_data_as_np.shape
```

```python
reshape_data = input_data_as_np.reshape(1, -1)
```

```python
prediction = model.predict(reshape_data)
if prediction == 0:
  print("The person is Non-Diabitic ")
else:
  print("The person is Diabetic")
```
